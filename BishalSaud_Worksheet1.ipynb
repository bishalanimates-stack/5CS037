```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

```python
# Task 1: Show A * A^-1 = I
A_inv = np.linalg.inv(A)
identity_check = A @ A_inv
identity_check
```

```python
# Task 2: Show AB != BA
AB = A @ B
BA = B @ A
AB, BA
```

```python
# Task 3: Verify (AB)^T = B^T A^T
lhs = (A @ B).T
rhs = B.T @ A.T
lhs, rhs
```

```python
# Task: Solve system AX=B using inverse and solve()

A = np.array([
    [2, -3, 1],
    [1, -1, 2],
    [3, 1, -1]
])

B = np.array([-1, -3, 9])

# Method 1: Using inverse
A_inv = np.linalg.inv(A)
X1 = A_inv @ B
X1
```

```python
# Method 2: Using np.linalg.solve()
X2 = np.linalg.solve(A, B)
X2
```

```python
import numpy as np
import time
```

```python
# Python addition
list1 = list(range(1_000_000))
list2 = list(range(1_000_000))

start = time.time()
py_add = [list1[i] + list2[i] for i in range(1_000_000)]
python_add_time = time.time() - start
python_add_time
```

```python
# NumPy addition
np1 = np.arange(1_000_000)
np2 = np.arange(1_000_000)

start = time.time()
np_add = np1 + np2
numpy_add_time = time.time() - start
numpy_add_time
```

```python
# Python multiplication
start = time.time()
py_mul = [list1[i] * list2[i] for i in range(1_000_000)]
python_mul_time = time.time() - start
python_mul_time
```

```python
# NumPy multiplication
start = time.time()
np_mul = np1 * np2
numpy_mul_time = time.time() - start
numpy_mul_time
```

```python
# Python dot product
start = time.time()
py_dot = sum(list1[i] * list2[i] for i in range(1_000_000))
python_dot_time = time.time() - start
python_dot_time
```

```python
# NumPy dot product
start = time.time()
np_dot = np.dot(np1, np2)
numpy_dot_time = time.time() - start
numpy_dot_time
```

```python
# Python matrix multiplication (very slow)
A_py = [[i for i in range(1000)] for _ in range(1000)]
B_py = [[i for i in range(1000)] for _ in range(1000)]

start = time.time()
C_py = [[sum(A_py[i][k] * B_py[k][j] for k in range(1000))
         for j in range(1000)] for i in range(1000)]
python_matmul_time = time.time() - start
python_matmul_time
```

```python
# NumPy fast matrix multiplication
A_np = np.arange(1_000_000).reshape(1000,1000)
B_np = np.arange(1_000_000).reshape(1000,1000)

start = time.time()
C_np = A_np @ B_np
numpy_matmul_time = time.time() - start
numpy_matmul_time
```
